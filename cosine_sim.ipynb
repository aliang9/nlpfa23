```python
!pip install pandas gensim scikit-learn
!pip install -U sentence-transformers
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

csv_path = '/content/drive/MyDrive/NLPProject/reddit-explanations-wgpt_filled.csv'
df = pd.read_csv(csv_path)


# Function to calculate cosine similarity between two strings given embeddings
def calculate_similarity(vec1, vec2):
    if np.linalg.norm(vec1) == 0 or np.linalg.norm(vec2) == 0:
        return 0.0  # Handle the case where a vector has zero norm
    return cosine_similarity([vec1], [vec2])[0][0]

column_pairs = [(2,3),
                (2,4),
                (2,5),
                (2,6),
                (2,7),
                (2,8)]

similarity_matrix = []

i = 0
for row in df.itertuples(index=False):
    if i < 6:
        i += 1
        continue
    similarities = [calculate_similarity(model.encode(row[col1]), model.encode(row[col2])) for col1, col2 in column_pairs]
    similarity_matrix.append(similarities)


import numpy as np
similarity_matrix = np.array(similarity_matrix)

print(similarity_matrix)
```

```python
average_values = np.mean(similarity_matrix, axis=0)
print(average_values)
```
