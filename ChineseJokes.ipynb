```python
import pandas as pd

# Specify the path to your Excel file (xlsx)
file_path_xlsx = '/content/Chinese_Humor_Multi-Labeled.xlsx'
# Read the Excel file into a DataFrame
df = pd.read_excel(file_path_xlsx, engine='openpyxl')

# Display the DataFrame
print("DataFrame from Excel file:")
print(df.head())
```

```python
columns_to_delete = ['ID', '標題', '雙關', '誇飾', '擬人化', '橋介推論', '法則誤用', '悖理', '模仿', '其他', '親和', '自我提升', '攻擊', '自我貶抑', '禁忌', '其他.1']
df = df.drop(columns=columns_to_delete, axis=1)
print(df.head())
```

```python
df['last_line'] = df['內容'].apply(lambda x: x.split('\n')[-1])

# Remove the last line from each cell in column A
df['內容'] = df['內容'].apply(lambda x: '\n'.join(x.split('\n')[:-1]))

# Display the DataFrame after the transformation
print("\nDataFrame after processing:")
print(df.head())
```

```python
columns = df.columns.tolist()

# Swap the last and second-to-last columns
columns[1], columns[2] = columns[2], columns[1]

# Reorder the DataFrame columns
df = df[columns]
print(df.head())
```

```python
column_to_filter = '好笑程度'  # Replace with the actual column name

# Filter the DataFrame for values greater than or equal to 2 in the specified column
df = df[df[column_to_filter] >= 2]

# Display the DataFrame after filtering
print("\nDataFrame after filtering values greater than or equal to 2:")
print(df.head())
num_rows = df.shape[0]
print(f"Number of rows: {num_rows}")

```

```python
output_file_path = '/content/updated-chinese-jokes.csv'

# Export the DataFrame to a new CSV file
df.to_csv(output_file_path, index=False)
```
